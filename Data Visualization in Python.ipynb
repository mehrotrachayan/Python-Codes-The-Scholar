```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
car=pd.read_csv('Car_Sales.csv')
```

```python
car.head(26)
```

```python
plt.figure(figsize=(10,10))
for x,y in car.groupby('Manufacturer'):
    plt.barh(x,y['4_year_resale_value'].mean())
```

```python
car['Vehicle_type'].unique()
```

```python
vt_dict=dict()
for i in set(car['Vehicle_type']):
    vt_dict[i]=car[car['Vehicle_type']==i].Manufacturer.count()
vt_dict
```

```python
plt.bar(vt_dict.keys(),vt_dict.values())
plt.show()
```

```python
plt.pie(vt_dict.values(),labels=vt_dict.keys())
plt.show()
```

```python
#Scatter Plot 
plt.scatter(car['Fuel_capacity'],car['Fuel_efficiency'])
```
